```python
%matplotlib inline
sns.set() #Use this line to plot the marginalized PDFs with seaborn style

plt.figure(figsize=(15, 10))
plt.suptitle('Marginalized PDFs')

plt.subplot(321)
sns.histplot( x='x', kde=True,data=Sample1, stat="density")
plt.title("PDF for X, sample 1")
plt.xlabel("X")
plt.ylabel("P(X)")

plt.subplot(322)
sns.histplot( x='x', kde=True,data=Sample1, stat="density")
plt.title("PDF for Y, sample 1")
plt.xlabel("Y")
plt.ylabel("P(Y)")

plt.subplot(323)
sns.histplot( x='x', kde=True,data=Sample2, stat="density")
plt.title("PDF for X, sample 2")
plt.xlabel("X")
plt.ylabel("P(X)")


plt.subplot(324)
sns.histplot( x='x', kde=True,data=Sample2, stat="density")
plt.title("PDF for Y, sample 2")
plt.xlabel("Y")
plt.ylabel("P(Y)")

plt.subplot(325)
sns.histplot( x='x', kde=True,data=Sample3, stat="density")
plt.title("PDF for X, sample 3")
plt.xlabel("X")
plt.ylabel("P(X)")


plt.subplot(326)
sns.histplot( x='x', kde=True,data=Sample3, stat="density")
plt.title("PDF for Y, sample 3")
plt.xlabel("Y")
plt.ylabel("P(Y)")


plt.tight_layout()
plt.show()


```
